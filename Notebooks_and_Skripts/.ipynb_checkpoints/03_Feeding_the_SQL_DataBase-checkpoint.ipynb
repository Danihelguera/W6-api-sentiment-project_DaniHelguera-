## 4. Iteramos fila por fila en todo el dataframe y para meter todas las filas en la BBDD

```python
ratings_que_dan_error = []
for index,fila in ars.iterrows() :
    #print(index)
    if index%100 == 0:
        print(f"Se han cargado ya {index} de un total de {len(ars)}")
    try: 
        insert_rating(fila);
    except Exception as err:
        print(f"Row {index} da error. Se registra.     --> {type(err).__name__} : {err}")
        ratings_que_dan_error.append( {"row" : index , "error_type" : type(err).__name__ , "error_message" : err.args } )
```
